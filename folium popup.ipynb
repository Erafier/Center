```python
import os
import folium

print(folium.__version__)
```

```python
icon_url = 'static/img/cctv.png'
icon_cctv = folium.features.CustomIcon(icon_url, icon_size=(32, 37))
coordinates = [
    (56.85702449459888, 60.65402140291714),
    (56.856225133114606, 60.65337767275356),
    (56.8572125771619, 60.65251936586883),
    (56.85582544594197, 60.65247645052457),
]


def draw_marker(map, coordinates, popup):
    for coord in coordinates:
        folium.Marker(coord, popup=popup).add_to(map)
```

```python
start_coords = (56.83906073448172, 60.611528966845654)
folium_map = folium.Map(location=start_coords, zoom_start=13)
draw_marker(map=folium_map, coordinates=coordinates, popup="Ghbdtn")
#folium.Marker((56.856225133114606, 60.65337767275356)).add_to(folium_map)
```
